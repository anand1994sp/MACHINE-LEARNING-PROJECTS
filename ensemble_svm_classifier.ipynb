```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/cancer.csv")
```

```python
df.drop("id",inplace=True,axis=1)
df.head(3)
```

```python
df.corr().T
```

```python
y=pd.DataFrame(df["diagnosis"])
```

```python
y.value_counts()
```

```python
x=df.drop("diagnosis",axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
y=pd.DataFrame(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.svm import SVC
svc=SVC(probability=True)
svc.fit(xtrain,ytrain)
```

```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
```

```python
ABC=AdaBoostClassifier(base_estimator=svc,n_estimators=100)
ABC.fit(xtrain,ytrain)
ypred=ABC.predict(xtest)
```

```python
from sklearn import metrics
```

```python
score=metrics.accuracy_score(ytest,ypred)
score
```

```python
gbc=GradientBoostingClassifier(n_estimators=100,max_depth=5)
gbc.fit(xtrain,ytrain)
```

```python
ypred=gbc.predict(xtest)
```

```python
score1=metrics.accuracy_score(ytest,ypred)
score1
```
